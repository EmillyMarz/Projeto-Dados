### O que representa cada coluna:

- `País`: Nome do país.
- `Ano`: Ano da observação.
- `Desenvolvimento`: Categoria de desenvolvimento (Desenvolvido/Desenvolvimento).
- `Expectativa_de_vida`: Expectativa média de vida (anos).
- `Mortalidade_adulta`: Mortalidade entre adultos (por 1000 habitantes).
- `Mortes_infantis`: Número de mortes infantis.
- `Álcool`: Consumo médio de álcool per capita.
- `Porcentagem_despesas`: Porcentagem de despesas com saúde.
- `Hepatite_B`: Percentual de vacinação contra Hepatite B.
- `Sarampo`: Número de casos de sarampo.
- `BMI`: Índice de massa corporal médio.
- `under-five_deaths`: Mortes abaixo de 5 anos.
- `Poliomielite`: Cobertura de vacinação contra poliomielite.
- `Despesa_total`: Despesa total com saúde.
- `Difteria`: Cobertura de vacinação contra difteria.
- `HIV/AIDS`: Casos registrados de HIV/AIDS.
- `PIB`: Produto Interno Bruto per capita.
- `População`: População total.
- `Magreza_1_a_19_anos`: Porcentagem de magreza em jovens.
- `Magreza_5_a_9_anos`: Porcentagem de magreza em crianças.
- `Composição_de_renda`: Índice de distribuição de renda.
- `Escolaridade`: Média de anos de escolaridade.
